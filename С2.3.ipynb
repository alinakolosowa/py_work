```python
import matplotlib.pyplot as plt 
s = [12692, 5393, 1494, 1195, 1157, 1100, 1095] 
x = range(len(s)) 
ax = plt.gca() 
plt.title("Численность  городов") 
ax.bar(x, s, align='edge') 
ax.set_xticks(x) 
ax.set_xticklabels(('Мск', 'С-Пб', 'Екб', 'Члб', 'Самара', 'Крснк', 'Воронеж')) 
plt.show()
```
